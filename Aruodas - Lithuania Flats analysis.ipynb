```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
df = pd.read_csv('aruodas.csv')
pd.set_option('display.max_columns', None)
```

```python
df.drop(df[df['Area m2'] > 500].index, inplace = True)
df.drop(df[df['Price EUR'] > 750000].index, inplace = True)
```

```python
df.drop(df[df['Specials'].str.contains('Varžytynės/aukcionas')].index, inplace = True)
```

```python
#After some cleaninig
df['City'].count()
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x='County',data=df,order=df['County'].value_counts().head(15).index);
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, fontsize=15);
ax.set_title("Average price by County top 15", fontsize=20)
```
